Y esto es un subtitulo que acompaña al título

```python
%display latex
x, c = var('x c')
f = 1/(x^2 + c)
g = derivative(f, x)
show('f(x) = ' + latex(f) + '\qquad f^\prime(x) =' + latex(g))
```

Y aquí abajo tenemos el plot de la función y el plot de la derivada

```python
W = plot(1/(x^2+ 1), (-1, 1))
S = plot(-2*x/((x^2+1)^2), (-1, 1))
show(W + S)
```
