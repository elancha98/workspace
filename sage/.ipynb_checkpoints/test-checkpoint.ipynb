```python
x, y = var('x y')

P = plot(sin(x), (x,0,pi), color='blue')
Q = plot(sin(x)+1, (x,0, pi), color='red', fill=True, fillcolor='red', fillalpha=0.4)
T = plot(0.935, (x, 0, pi), color='black',fill=True)
H = plot(sin(x)+1/pi, (x, 0, pi), color='green')
O = plot(sin(x)-1/pi, (x,0, pi), color='green')
p = point((0, -1))
q = point((0, 3))
show(P+T+Q+H+O+p+q)
```

```python
def g(t):
    if (t <= 0.665):
        return 0.935
    elif (t >= 2.477):
        return 0.935
    else:
        return sin(t)+1/pi
    
    
V = plot(g(x), (x, 0, pi))
l = [(x*pi/1000, g(x*pi/1000)) for x in range(0, 1000, 1)]
R = list_plot(l)
show(R+V)
```
