```
x, y = var('x y')

P = plot(sin(x), (x,0,pi), color='blue')
T = plot(0.935, (x, 0, pi), color='black')
H = plot(sin(x)+1/pi, (x, 0, pi), color='green')
O = plot(sin(x)-1/pi, (x,0, pi), color='green')
p = point((0, -1))
q = point((0, 3))
show(P+T+H+O+p+q)
```

```
def g(t):
    if (t <= 0.665):
        return 0.935
    elif (t >= 2.477):
        return 0.935
    else:
        return sin(t)+1/pi
    
    
V = plot(g(x), (x, 0, pi))
l = [(x*pi/1000, g(x*pi/1000)) for x in range(0, 1000, 1)]
R = list_plot(l)
show(R+V)
```
